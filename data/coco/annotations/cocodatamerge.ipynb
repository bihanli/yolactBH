create categories image and dump json

```
import json
import base64
import numpy as np
import os
input_file = "/content/drive/My Drive/instances_train2017_person.json"
dic={}
dic["images"]=[]
dic["annotation"]=[]
with open (input_file,'r')as f:
  dd=json.load(f)

u=[1,2,3,4,5,6,7,8,10,11,13,14,16,17,18]
for i in u:
  count=0
  for j,k in enumerate(dd["annotations"]):
    if k["category_id"]==u:
      count+=1
      if count<=50:
        dic["annotation"].append(k)
      else:
        break
  print(count)
for l in dic["annotation"]:
  img=l["image_id"]
  image="%012d"%img
  print(image)
  for m in dd["images"]:
    t=str(image)+".jpg"
    if m["file_name"]==t and m not in dic["images"]:
      dic["images"].append(m)
dic["annotations"]=[]
for k in dic["images"]:
  for o in dd["annotations"]:
    if k["id"]==o["image_id"] and o["category_id"] in u:
      dic["annotations"].append(o)
del dd["images"]
del dd["annotations"]
del dic["annotation"]
dd["images"]=dic["images"]
dd["annotations"]=dic["annotations"]
with open ('/content/drive/MyDrive/yolact-master/data/coco/2017_NEW.json','w') as ff:    
  json.dump(dd,ff,sort_keys=True,indent=4)
print("END")
```

coco json image load into folder

```
import shutil
import json
coco_file="/content/drive/MyDrive/yolact-master/data/coco/2017_NEW.json"    
with open(coco_file,'r') as co:
  coco=json.load(co)
for i in coco["images"]:
  img=i["file_name"]
  print(img)
  img_path="/content/drive/MyDrive/yolact-master/data/coco/train2017/"+str(img)
  output="/content/drive/MyDrive/yolact-master/data/coco/train/"+str(img)
  shutil.copyfile(img_path,output)
print("END")
```

```
import os
end="jpg"
folder="/content/drive/MyDrive/yolact-master/data/coco/train/" 
allfile=os.listdir(folder)
for image in allfile:
  if image.endswith(end):
    os.remove(image)
```
