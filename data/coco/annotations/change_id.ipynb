```
import json
with open('/content/drive/MyDrive/yolact-master/data/coco/annotations/LabelWY.json','r') as f:
  file=json.load(f)
  
  for i in file["categories"] :
    if i["name"]=="traffic light":
      i["supercategory"]="outdoor"
      i["id"]=10
    elif i["name"]=="motorcycle":
      i["supercategory"]="vehicle"
      i["id"]=4
  for i in file["annotations"]:
    if i["category_id"]==4:
      i["category_id"]=10
    elif i["category_id"]==2:
      i["category_id"]=4
  print(file)
  print(file["categories"])
with open('/content/drive/MyDrive/yolact-master/data/coco/annotations/LabelWY.json','w') as ff:
  json.dump(file,ff,sort_keys=True,indent=4)
```
